```python
import numpy as np
import pandas as pd 
import folium
import folium.plugins
import json
import io
```

```python
# read files 
by_state = pd.read_csv("Earnings_By_State.csv")
by_county = pd.read_csv("Earnings_By_County.csv")

counties = ("gz_2010_us_050_00_5m.json")
state = ("gz_2010_us_040_00_500k.json")
```

```python
# rename column back to state 
by_state = by_state.rename(columns = {"Unnamed: 0": "State"})

# Split the name into both county and state so they can be grouped later
by_county[["County", "State"]] = by_county.NAME.str.split(",", expand = True)
```

```python
# to fix the UnicodeDecodeError from how the geojson file is encoded
with io.open(counties, encoding="latin1") as file:
    county_geo = json.load(file)

with open(state, "r") as file:
    state_geo = json.load(file)
```

```python
# change District of Coluumbia in the df to District of Columbia so it's the same as geojson file
by_state = by_state.replace("District of Coluumbia", "District of Columbia")
```

```python
# delete county and parish from the name so it can match with the geojson 
by_county["County"] = by_county["County"].str.replace(" County", "")
by_county["County"] = by_county["County"].str.replace(" Parish", "")
```

### Maps For States

```python
# map for total state median income
s = folium.Map(location=(39, -98), zoom_start=4, tiles="cartodb positron")

folium.Choropleth(
    geo_data=state,
    data=by_state,
    columns=("State", "Total"),
    key_on="feature.properties.NAME", # keys to link the data with gejson
    # bins are created depending on the range of the income for each column
    bins=(30_000, 35_000, 40_000, 45_000, 50_000, 55_000, 60_000, 70_000, 80_000, by_state["Total"].max()),
    fill_color="YlGn",
    fill_opacity=0.8,
    line_opacity=0.3,
    legend_name="Total Median Income",
    name="Counties by Total Median Income",
).add_to(s)

folium.LayerControl().add_to(s)
fig = folium.Figure(width = 800, height = 500)

s.save("us_total_median_income.html")
```
